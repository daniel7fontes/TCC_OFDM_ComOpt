# DDO-OFDM

## 1 - Transmissor

```python
paramTx = parameters()

# Parâmetros da transmissão
paramTx.SpS = 32                         # Amostras por símbolo
paramTx.Rs  = 1.5e9                      # Taxa de símbolo
paramTx.Ts  = 1/paramTx.Rs               # Tempo de símblo
paramTx.Fa  = 1/(paramTx.Ts/paramTx.SpS) # Frequência de amostragem
paramTx.Ta  = 1/paramTx.Fa               # Período de amostragem
paramTx.Scheme = "DD-OFDM-Real"

# Parâmetros do esquema OFDM
paramTx.M = 4                            # Número de símbolos da constelação QAM
paramTx.N = 512                          # Dobro do número de sub-portadoras
paramTx.G = 16                           # Tamanho do prefixo cíclico
paramTx.K = 16                           # Número de portadoras piloto por bloco OFDM

# Parâmetros da portadora elétrica
paramTx.H  = 0.35/(2*pi)                 # Índice de modulação
paramTx.fc = 1e9                         # Frequência da portadora [Hz]
paramTx.A  = 1                           # Amplitude da portadora

# Parâmetros do MZM e potência de entrada
paramTx.Vπ     = 4.7
paramTx.Vb     = -0.7*paramTx.Vπ
paramTx.Pi_dBm = 8
```

```python
sigTxo, sigTx, sigTx_DD, symbTx, t, pulse, pilot = Tx(paramTx)
```

#### Sinais no domínio do tempo

```python
fig, ax = plt.subplots(1, 3, figsize = (24, 6))
idX = np.arange(0, 1024)

ax[0].plot(t[idX]*1e9, sigTx[idX].real)
ax[0].set_xlabel('Tempo [ns]')
ax[0].set_ylabel('Amplitude [u.a.]')
ax[0].set_title('SigTx')
ax[0].grid()

ax[1].plot(t[idX]*1e9, sigTx_DD[idX].real)
ax[1].set_xlabel('Tempo [ns]')
ax[1].set_ylabel('Amplitude [u.a.]')
ax[1].set_title('SigTx_DD')
ax[1].grid()

ax[2].plot(t[idX]*1e9, sigTxo[idX].real)
ax[2].set_xlabel('Tempo [ns]')
ax[2].set_ylabel('Amplitude [u.a.]')
ax[2].set_title('SigTxo')
ax[2].grid()
```

#### Espectros

```python
fig, ax = plt.subplots(1, 3, figsize = (24, 6))
idX = np.arange(0, 512)

ax[0].psd(sigTx, Fs=paramTx.Fa, NFFT = 16*1024, sides='twosided')
ax[0].set_title('SigTx')
ax[0].set_xlim(-5e9, 5e9)
ax[0].grid()

ax[1].psd(sigTx_DD, Fs=paramTx.Fa, NFFT = 16*1024, sides='twosided')
ax[1].set_title('SigTx_DD')
ax[1].set_xlim(-5e9, 5e9)
ax[1].grid()

ax[2].psd(sigTxo, Fs=paramTx.Fa, NFFT = 16*1024, sides='twosided')
ax[2].set_title('SigTxo')
ax[2].set_xlim(-20e9, 20e9)
ax[2].set_ylim(-250, -90)
ax[2].grid()
```

```python
print(f" SigTx    - Pi[dBm] = {10*np.log10(1000*signal_power(sigTx))}")
print(f" SigTx_DD - Pi[dBm] = {10*np.log10(1000*signal_power(sigTx_DD))}")
print(f" SigTxo   - Pi[dBm] = {10*np.log10(1000*signal_power(sigTxo))}")
```

```python
# PAPR
L = 1000
PP = max(sigTx_DD[0:L*(paramTx.N+paramTx.G)].real)**2/signal_power(sigTx_DD[0:L*(paramTx.N+paramTx.G)].real)

PAPR = 10*np.log10(PP)
print('PAPR = %.2f dB' %(PAPR))
```

## 2 - Canal óptico

```python
paramCh = parameters()

# Parâmetros da fibra
paramCh.Ltotal = 400         # Comprimento total do link [km]
paramCh.Lspan  = 50          # Comprimento de um span de fibra [km]
paramCh.hz     = 2           # Comprimento do passo para o método SSF [km]
paramCh.alpha  = 0.2         # Atenuação da fibra [dB/km]
paramCh.D      = 16          # Parâmetro de dispersão cromática [ps/nm/km]
paramCh.gamma  = 1.3         # Parâmetro de não-linearidade da fibra [1/W/km]
paramCh.amp    = 'edfa'      # Tipo de amplificador
paramCh.NF     = 5.0         # Figura de ruído do amplificador [dB]
paramCh.Fc     = paramTx.Fc  # Frequência da portadora óptica [Hz]
paramCh.prgsBar = True
```

```python
sigRxo,_  = ssfm(sigRxo, paramTx.Fa, paramCh)
```

## 3 - Recepção

```python
paramPD = parameters()

# Parâmetros do fotodetector
paramPD.R  = 1           # Responsividade
paramPD.Tc = 25          # Temperatura [°C]
paramPD.Id = 5e-9        # Corrente de escuro [A]
paramPD.RL = 50          # Resistência [Ω]
paramPD.B  = 10e9        # Largura de banda [Hz]
paramPD.Fa = paramTx.Fa  # Frequência de amostragem
paramPD.N  = 8001        # Tamanho do resposta do filtro
paramPD.ideal = False    # Definição do modelo: ideal ou real
```

```python
ipd = photodiode(sigRxo, paramPD)
```

```python
paramRx = parameters()

# Parâmetros do receptor
paramRx.SpS = paramTx.SpS       # Amostras por símbolo
paramRx.Fa  = paramTx.Fa        # Frequência de amostragem
paramRx.H   = paramTx.H         # Índice de modulação
paramRx.fc  = paramTx.fc        # Frequência da portadora elétrica
paramRx.Scheme = paramTx.Scheme

paramRx.N = paramTx.N
paramRx.G = paramTx.G
paramRx.K = paramTx.K
```

```python
symbRx, symbRx_neq, H_abs, H_pha, I_Rx = Rx(ipd, pilot, pulse, t, paramRx)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (16, 5))

ax[0].plot(H_abs)
ax[0].set_title("|H| channel estimation")
ax[0].set_xlabel("Sub-portadora")
ax[0].grid()

ax[1].plot(H_pha*180/pi)
ax[1].set_title(r"$\angle H$ channel estimation")
ax[1].set_xlabel("Sub-portadora")
ax[1].grid()
```

#### Sinal óptico recebido

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 6))
idX = np.arange(0, 512)

ax[0].plot(t[idX]*1e9, np.abs(sigTxo[idX]), label = "SigTxo")
ax[0].plot(t[idX]*1e9, np.abs(sigRxo[idX]), label = "SigRxo")
ax[0].set_xlabel('Tempo [ns]')
ax[0].set_ylabel('Amplitude [u.a.]')
ax[0].grid()

ax[1].psd(sigTxo, Fs=paramTx.Fa, NFFT = 16*1024, sides='twosided', label = "SigTxo")
ax[1].psd(sigRxo, Fs=paramTx.Fa, NFFT = 16*1024, sides='twosided', label = "SigRxo")
ax[1].set_xlim(-20e9, 20e9)
```

## 4 - Desempenho do sistema

### 4.1 - Com equalização

```python
symbRx[0::int(paramTx.N/2)] = 'nan'
symbTx[0::int(paramTx.N/2)] = 'nan'

symbTx = symbTx[~np.isnan(symbTx)]
symbRx = symbRx[~np.isnan(symbRx)]

# Cálculo da BER
np.log10(fastBERcalc(symbRx, symbTx, paramTx.M, 'qam')[0])
```

```python
# Plot da constelação transmitida e recebida
plt.figure(figsize=(8, 8))
plt.plot(symbRx.real, symbRx.imag, '.', label = 'Rx')
plt.plot(symbTx.real, symbTx.imag, '.', label = 'Tx')
plt.axis('square')
plt.xlabel('In-Phase (I)')
plt.ylabel('Quadrature (Q)')
plt.legend(loc = 'upper left', frameon = True)
plt.tight_layout()
plt.grid()

plt.xlim(-2,2)
plt.ylim(-2,2)
```

### 4.2 - Sem equalização

```python
symbRx_neq[0::int(paramTx.N/2)] = 'nan'

symbRx_neq = symbRx_neq[~np.isnan(symbRx_neq)]

# Cálculo da BER
np.log10(fastBERcalc(symbRx_neq, symbTx, paramTx.M, 'qam')[0])
```

```python
# Plot da constelação transmitida e recebida
plt.figure(figsize=(8, 8))
plt.plot(symbRx_neq.real, symbRx_neq.imag, '.', label = 'Rx')
plt.plot(symbTx.real, symbTx.imag, '.', label = 'Tx')
plt.axis('square')
plt.xlabel('In-Phase (I)')
plt.ylabel('Quadrature (Q)')
plt.legend(loc = 'upper left', frameon = True)
plt.tight_layout()
plt.grid()

plt.xlim(-2,2)
plt.ylim(-2,2)
```

```python
# CE-DD-OFDM com passo hz = 2
Pi_dBm = np.array([2, 8])
BER_DD2 = np.zeros( (len(Pi_dBm), 20) )

paramTx.Scheme = "DD-OFDM-Real"
paramRx.Scheme = "DD-OFDM-Real"
paramCh.prgsBar = False
paramCh.hz = 2

for i in tqdm(range(len(Pi_dBm))):
    paramTx.Pi_dBm = Pi_dBm[i]
    paramCh.Ltotal = 100
    sigTxo, sigTx, sigSig, symbTx, t, pulse, pilot = Tx(paramTx)
    sigRxo = sigTxo
    
    symbTx[0::int(paramTx.N/2)] = 'nan'
    symbTx = symbTx[~np.isnan(symbTx)]
    
    for j in range(20):
        sigRxo,_  = ssfm(sigRxo, paramTx.Fa, paramCh)
        ipd = photodiode(sigRxo, paramPD)
        symbRx, symbRx_neq, H_abs, H_pha, I_Rx = Rx(ipd, pilot, pulse, t, paramRx)
        
        symbRx[0::int(paramTx.N/2)] = 'nan'
        symbRx = symbRx[~np.isnan(symbRx)]

        # Cálculo da BER
        BER_DD2[i, j] = fastBERcalc(symbRx, symbTx, paramTx.M, 'qam')[0]
```

```python
L = np.arange(100, 2100, 100)
plt.figure(figsize = (8, 6))

for i in range(2):
    plt.plot(L, np.log10(BER_DD2[i]), "o-",label = f"Pi = {Pi_dBm[i]} dBm")
    #plt.plot(L, np.log10(BER_DD[i]), "^-")
    plt.legend()

plt.xlabel("L [km]")
plt.ylabel(r"$\log _{10}(BER)$")
plt.ylim(-5, 0)
```
