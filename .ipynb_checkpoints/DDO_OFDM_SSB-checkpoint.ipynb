# DDO-OFDM

```python
from scipy.fftpack     import fft, ifft
from scipy.constants   import pi
from scipy.signal      import hilbert
from scipy.interpolate import interp1d

from optic.modulation  import modulateGray, demodulateGray, GrayMapping
from optic.dsp         import pulseShape, lowPassFIR, firFilter
from optic.metrics     import signal_power
from optic.models      import mzm
```

```python
M    = 4
Ns   = 12           # Sub-portadoras úteis
Nfft = 32
N    = int(Nfft/2 - 1)  
Nz   = N - Ns        # Número de sub-portadoras preenchidas com 0

G = 4
K = 1
SpS = 32

aux = 0
fc  = 1e9

Rs  = 1.50e9
Ts  = 1/Rs             # Symbol period
Fa  = 1/(Ts/SpS)       # Sampling frequency
Ta  = 1/Fa             # Sampling period 
```

```python
N
```

```python
((Nfft)*Ts)*1e9 # Tempo de símbolo OFDM [ns]
```

```python
G*Ts*1e9 # Duração do prefixo cíclico
```

```python
# Symbols constellation definiton
constType = 'qam'
constSymb = GrayMapping(M, constType)
bitMap = demodulateGray(constSymb, M, constType)
bitMap = bitMap.reshape(-1, int(np.log2(M)))

# Random bits sequency
bits = np.random.randint(2, size = Ns*2**10)
    
# Maping bits - symbols
symbTx = modulateGray(bits, M, constType)
symbTx = symbTx/np.sqrt(signal_power(symbTx))

# Pilot carriers
pilot = max(symbTx.real) + 1j*max(symbTx.imag)
```

```python
Int = int(Ns/K)
carriers = np.arange(0, Ns)
pilot_carriers = np.append(carriers[0::Int], carriers[-1])
  
symbTx_P = np.zeros( (int(len(symbTx)/Ns), Nfft + G), complex)
aux = 0
for i in range(len(symbTx_P)):
    # Adição das portadoras piloto
    symbTx[int(aux):int(aux + Ns)][pilot_carriers] = pilot
    
    symbTx_P[i, G : G + Nfft] = hermit( np.concatenate( (np.zeros(int(Nz)), symbTx[int(aux):int(aux + Ns)]) ) )
    aux = aux + Ns
    
    if(i == 0):
        aa1 = symbTx_P[0].copy()
    
    # Aplicação da IFFT
    symbTx_P[i, G : Nfft + G] = ifft(symbTx_P[i, G:Nfft+G])*np.sqrt(Nfft)
    
    if(i == 0):
        aa2 = symbTx_P[0].copy()
    
    # Adição do prefixo cíclico
    symbTx_P[i,:] = np.concatenate((symbTx_P[i,Nfft:Nfft+G],symbTx_P[i,G:Nfft+G]))

# Conversão Paralelo -> Serial
symbTx_OFDM = np.squeeze(symbTx_P.reshape(1,len(symbTx_P[0])*len(symbTx_P)))
```

```python
aa1
```

```python
pulse = pulseShape('rrc', SpS, alpha = 0.15)
pulse = pulse/max(abs(pulse))

sigTx = firFilter(pulse, upsample(symbTx_OFDM.real, SpS))
t = np.arange(0, sigTx.size)*Ta

Ai = np.sqrt(Pi) * np.ones(len(sigTx))
sigTx_CE = np.cos(2*pi*fc*t + 0.3*sigTx.real/np.std(sigTx.real))
sigTxo = mzm(Ai, sigTx_CE, 4, -2)
```

```python
M    = 4
Nfft = 512
Ns   = int(Nfft/4)
N    = Nfft/2 - 1  

G = 4
K = 2
SpS = 32

Rs  = 2.5e9
Ts  = 1/Rs             # Symbol period
Fa  = 1/(Ts/SpS)       # Sampling frequency
Ta  = 1/Fa             # Sampling period   

# DDO-OFDM III (SSB)
aux = 0

symbTx_P = np.zeros( (int(len(symbTx)/Ns), Nfft + G), complex)

for i in range(len(symbTx_P)):
    # Adição das portadoras piloto
    #symbTx[int(aux):int(aux + Ns)][pilot_carriers] = pilot
    
    symbTx_P[i, G : G + Nfft] = np.concatenate( (np.zeros(int(Ns)), symbTx[int(aux):int(aux + Ns)], np.zeros(int(2*Ns))) )
    aux = aux + Ns
    
    # Aplicação da IFFT
    symbTx_P[i, G : Nfft + G] = ifft(symbTx_P[i, G:Nfft+G])*np.sqrt(Nfft)
    
    # Adição do prefixo cíclico
    symbTx_P[i,:] = np.concatenate((symbTx_P[i,Nfft:Nfft+G],symbTx_P[i,G:Nfft+G]))

# Conversão Paralelo -> Serial
symbTx_OFDM = np.squeeze(symbTx_P.reshape(1,len(symbTx_P[0])*len(symbTx_P)))
```

```python
symbTx_OFDM
```

```python
def hermit(V):
    Vh = np.zeros(2*len(V) + 2, complex)
    
    Vh[1:len(V)+1] = V 
    
    for j in range(len(V)):
        Vh[len(Vh) - j - 1] = np.conjugate(V[j])
    
    return Vh
```

```python
pulse = pulseShape('rrc', SpS, alpha = 0.15)
pulse = pulse/max(abs(pulse))

#sigTx_I = firFilter(pulse, upsample(symbTx_OFDM.real, SpS))
#sigTx_Q = firFilter(pulse, upsample(symbTx_OFDM.imag, SpS))
sigTx = firFilter(pulse, upsample(symbTx_OFDM.real, SpS))
t = np.arange(0, sigTx.size)*Ta

Ai = np.sqrt(Pi) * np.ones(len(sigTx))
sigTx_CE = np.cos(2*pi*fc*t + 0.1*sigTx/np.std(sigTx))
sigTxo = mzm(Ai, sigTx_CE, 4, -2)

# Optical modulation
#sigTx_DD = np.cos(2*pi*fc*t + 0.35*sigTx.real/np.std(sigTx.real)) #sigTx.real*np.cos(2*pi*fc*t)



#sigTxo = iqm(Ai, sigTx_CE_I + 1j*sigTx_CE_Q, 4, -2, -2)
#sigSig = sigTx_DD
```

## 3 - Recepção

```python
paramPD = parameters()

# Parâmetros do fotodetector
paramPD.R  = 1           # Responsividade
paramPD.Tc = 25          # Temperatura [°C]
paramPD.Id = 5e-9        # Corrente de escuro [A]
paramPD.RL = 50          # Resistência [Ω]
paramPD.B  = 10e9        # Largura de banda [Hz]
paramPD.Fa = Fa          # Frequência de amostragem
paramPD.N  = 8001        # Tamanho do resposta do filtro
paramPD.ideal = False    # Definição do modelo: ideal ou real
```

```python
ipd = photodiode(sigRxo, paramPD)
```

```python
paramRx = parameters()

# Parâmetros do receptor
paramRx.SpS = paramTx.SpS       # Amostras por símbolo
paramRx.Fa  = Fa                # Frequência de amostragem
paramRx.H   = paramTx.H         # Índice de modulação
paramRx.fc  = paramTx.fc        # Frequência da portadora elétrica
paramRx.Scheme = paramTx.Scheme

paramRx.N = paramTx.N
paramRx.G = paramTx.G
paramRx.K = paramTx.K
```

```python
from scipy.signal import hilbert
```

```python
I_Rx = ipd - ipd.mean()
I_Rx = I_Rx#/np.std(I_Rx)

signal_a = firFilter(pulse/SpS, hilbert(I_Rx) * np.exp(-1j*2*pi*fc*t))
Θ        = (np.arctan(signal_a.imag/signal_a.real))/(0.35) 

symbRx_OFDM = Θ[0::SpS]
```

```python
#Ns = int(N/2)
#Int = int(Ns/K) 
#carriers = np.arange(0, Ns)
#pilot_carriers = np.append(carriers[1::Int], carriers[-1])
    
# Recebendo um sinal OFDM real
# Conversão S/P
symbRx_OFDM = symbTx_OFDM.copy()
symbRx_P = np.zeros((int(len(symbRx_OFDM)/(Nfft+G)), Nfft), complex)

aux = 0
H_abs_F = 0
H_pha_F = 0

for i in range(len(symbRx_P)):
    # Extração do prefixo cíclico
    symbRx_P[i,0:Nfft] = symbRx_OFDM[aux+G:aux+Nfft+G]
    aux = aux + Nfft + G

    # Aplicação da FFT
    symbRx_P[i,:] = fft(symbRx_P[i,:])/np.sqrt(Nfft)

for i in range(len(symbRx_P)):
    H_est = symbRx_P[i,1 + Nz:1 + Nz + Ns][pilot_carriers] / pilot
    
    H_abs = interp1d(carriers[pilot_carriers], np.abs(H_est), kind = 'linear')(carriers)
    H_pha = interp1d(carriers[pilot_carriers], np.angle(H_est), kind = 'linear')(carriers)

    H_abs_F += H_abs
    H_pha_F += H_pha

H_abs = H_abs_F/len(symbRx_P)
H_pha = H_pha_F/len(symbRx_P)
    
# Conversão P/S
symbRx_S     = np.zeros(int(len(symbRx_P)*Ns), complex)  # Símbolos equalizados
symbRx_S_neq = np.zeros(int(len(symbRx_P)*Ns), complex)  # Símbolos não-equalizados
aux = 0

for i in range(len(symbRx_P)):
    # Retirada da simetria hermitiana e equalização
    symbRx_S[int(aux):int(aux + Ns)]     = symbRx_P[i,1 + Nz:1 + Nz + Ns]
    symbRx_S_neq[int(aux):int(aux + Ns)] = symbRx_P[i,1 + Nz:1 + Nz + Ns]/(H_abs*np.exp(1j*H_pha))
    aux = aux + Ns

symbRx_eq  = symbRx_S
symbRx_neq = symbRx_S_neq
```

```python
symbRx, symbRx_neq, H_abs, H_pha, I_Rx = Rx(ipd, pilot, pulse, t, paramRx)
```
